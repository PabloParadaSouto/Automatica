En este tercer y último ejemplo entrenamos una red con tres capas, dos ocultas de 128 neuronas cada una, y la tercera y última capa de salida de 10 neuronas (igual que en el caso anterior).

```python
#En primer lugar cargamos librerias que necesitamos (por eso estas lineas empiezan por la palabra "import"
import tensorflow as tf
import numpy as np
from tensorflow import keras
from matplotlib import pyplot as plt
from IPython.display import clear_output

#cargamos la base de datos MNIST que contiene los digitos
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

#vamos comprobar ahora la dimension de nuestros datos
#x_train --> representan las entradas que voy a usar para entrenar (train) la red
#y_train --> representa la salida deseada para cada uno de los datos en x_train
print ("dimension del array para entrenamiento ",x_train.shape)
print ("filas ",x_train.shape[0])
print ("cada fila tiene un patron bidimensional de  ",x_train.shape[1]," x ",x_train.shape[2])

#Si ejecutas este trozo de codigo, veras que te muestra la siguiente informacion:
# dimension del array  para entrenamiento (60000, 28, 28) --> el primer numero 60000 indica el numero de entradas,
#los dos numeros siguientes ,28,28) indican que cada entrada es una matriz 28x28, esto es, una imagen de 28 x 28
#pixeles


#definimos valores importantes
#estos valores no se deben modificar...
numero_clases=10
#convertimos las imagenes de 28x28 en un vector de 784 componentes
reshaped=784

#estos parametros definen como va a funcionar la red
epocas=200
tamanho_batch=128
VERBOSE=1
numero_ocultas=128
VALIDATION_SPLIT=0.2

probabilidad_dropout=0.3

#recordemos que para entrenamiento tenemos 60000 muestras
x_train=x_train.reshape(60000,reshaped)
x_test=x_test.reshape(10000,reshaped)

x_train=x_train.astype('float32')
x_test=x_test.astype('float32')

#normalizamos las entradas, las imagenes, pixeles, son valores entre 0 y 255. Ahora los pasamos a valores entre 0 y 1
x_train/=255
x_test/=255

print (x_train.shape[0])
print (x_test.shape[0])

#pasamos a la one-hot representation para las etiquetas
y_train=tf.keras.utils.to_categorical(y_train,numero_clases)
y_test=tf.keras.utils.to_categorical(y_test,numero_clases)
print ("despues del one-hot encoding las nueas dimensiones ")
print ("etiquetas entrenamiento ",y_train.shape)
print ("etiquetas test ",y_test.shape)

#Creamos la red de neuronas, en este caso estaría formado por una única capa:
model=tf.keras.models.Sequential()
#definimos el nuemro de salidas: numero_clases
#definimos como son las entradas=input_shape(reshaped,)
#Creamos una primera capa de neuronas ocultas (128 neuronas ocultas)
model.add(keras.layers.Dense(numero_ocultas,input_shape=(reshaped,),name='capa_oculta1',activation='relu'))

model.add(keras.layers.Dense(numero_ocultas,name='capa_oculta2',activation='relu'))
#Al igual que en el ejemplo anterior anhadimos ahora la capa de salida de 10 neuronas (una por clase)
model.add(keras.layers.Dense(numero_clases,name='capa_de_salida',activation='softmax'))

#compilamos el modelo (Definimos algoritmo de optimización SGD=stochastic gadient descent)
model.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])

#otras funciones de perdida (loss) podrian ser:
#binary_crossentropy
#mse
#podeis averiguar mas sobre estas otras opciones

#Entrenamos ahora el modelo
model.fit(x_train,y_train,batch_size=tamanho_batch,epochs=epocas,verbose=VERBOSE,validation_split=VALIDATION_SPLIT)

#evaluamos el modelo
test_loss,test_acc=model.evaluate(x_test,y_test)
print ('test accuracy ',test_acc)

```
